# Dealing with missing values
- in pandas,missing data is represented by two values:
- None : used for missing data in python code
- NaN : floating-point value recognized by all systems that use standard IEEE floating-point representation



# Pandas isnull() and notnull() functions

Functions to test for missing data-1.isnull() 2.notnull()
- These are simple functions that return a boolean value indicating whether passed in argument value is in fact missing data

# Commands to detect missing values
- df.isnull() : checks whether each cell in dataframe contains missing value or not
- df.isnull().sum() : Returns total no. of missing values in each column in dataframe
- df.isnull().sum().sum() : returns total no. of missing values in dataframe
- df.isnull().mean() : returns percentage of missing values ineach column in dataframe
- df.isnull().any() : checks which column has null values and which has not.(True and False)
- df.isnull().any().any() : returns boolean value indicatingg whether dataframe has missing values or not
- df.isnull().values.any() : it checks whether particular column has missing value or not,returns boolean value
- df.isnull().values.sum() : returns total no. of missing values in dataframe

```python
df.isnull().sum() 
```

- There is no missing value in dataset

# Assert Statement
- check that no missing,unexpected 0 or negative values are present.
- It will return nothing if value is being tested is TRUE and will throw an AssertionError if value is False
- Asserts
- assert 1 == 1 (return nothing if value is True)
- assert 1 == 2 (return AssertionError if value is False)

```python
# asserts that there are no missing values in dataframe

assert pd.notnull(df).all().all()
```

```python
# assert all values are greater than or equal to 0

assert(df>=0).all().all()
```

- Above 2 commands do not throw any error 
- i.e. there are no missing/negative values in dataset
- all values are greater than or equal to 0
